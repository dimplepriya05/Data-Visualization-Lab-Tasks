```python
import networkx as nx
import matplotlib.pyplot as plt

sales_network = nx.Graph()
sales_network.add_nodes_from(["Product A", "Product B", "Store 1", "Store 2", "Store 3"])
sales_network.add_edges_from([
    ("Product A", "Store 1"), ("Product A", "Store 2"),
    ("Product B", "Store 2"), ("Product B", "Store 3"),
    ("Store 1", "Store 2"), ("Store 2", "Store 3")
])

pos = nx.spring_layout(sales_network)
nx.draw(sales_network, pos, with_labels=True, node_size=1000, node_color='skyblue', font_size=10)
plt.show()
```
