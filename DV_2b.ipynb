```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/2b_dataset.csv")

categorical_cols = df.select_dtypes(include=['object', 'category']).columns
continuous_cols = df.select_dtypes(include=['int64', 'float64']).columns

print("Categorical Attributes:", categorical_cols.tolist())
print("Continuous Attributes:", continuous_cols.tolist())

cat_col = categorical_cols[0] if len(categorical_cols) > 0 else None
cont_col = continuous_cols[0] if len(continuous_cols) > 0 else None

if cat_col:
    plt.figure(figsize=(6,4))
    df[cat_col].value_counts().plot(kind='bar', color='skyblue')
    plt.title(f"Bar Chart of {cat_col}")
    plt.xlabel(cat_col)
    plt.ylabel("Count")
    plt.show()


    plt.figure(figsize=(6,6))
    df[cat_col].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette("pastel"))
    plt.title(f"Pie Chart of {cat_col}")
    plt.ylabel("")
    plt.show()


if cont_col:

    plt.figure(figsize=(6,4))
    sns.histplot(df[cont_col], bins=20, kde=False, color='orange')
    plt.title(f"Histogram of {cont_col}")
    plt.xlabel(cont_col)
    plt.ylabel("Frequency")
    plt.show()


    plt.figure(figsize=(6,4))
    sns.kdeplot(df[cont_col], shade=True, color='green')
    plt.title(f"Density Plot of {cont_col}")
    plt.xlabel(cont_col)
    plt.ylabel("Density")
    plt.show()


    plt.figure(figsize=(6,4))
    sns.rugplot(df[cont_col], color='red')
    plt.title(f"Rug Plot of {cont_col}")
    plt.xlabel(cont_col)
    plt.show()
```
